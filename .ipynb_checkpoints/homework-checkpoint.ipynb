Значительная часть различих объясняется тем, что mystem возвращает все глаголы в формы НСВ, а pymorphy сохраняет формы  CВ
Ещё, pymorphy использует букву "ё"
они еще и отличаются в формах прилагательных и наречия - например, в pymorphy активный, тёплый, и большой остаются прилагательными, а в mystem они наречия, но в mystem 'нужно', наоборать, превращается в 'нужный'
Ещё есть маленкие различия: деньга /деньги, так/т это/этот больший/большой сам/cамый

Если сохранение формы СВ или отличия между прилагательными и наречиями важны для конкретной задачи, тогда pymoprhy лучше. Если, наоборот, только основном значение слова важно, тогда mystem лучше
